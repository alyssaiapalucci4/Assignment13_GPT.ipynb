```python
# Import Libraries:

import torch
import requests
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

### Dataset Preparation

We use a public domain dataset from Project Gutenberg: *Alice’s Adventures in Wonderland*. It is suitable for text generation because it contains structured narrative language.

```python
# Load dataset from Project Gutenberg

url = "https://www.gutenberg.org/files/11/11-0.txt"
response = requests.get(url)
text = response.text

print("Dataset loaded successfully")
print("Total characters:", len(text))
print("\nPreview:\n")
print(text[:500])
```

### GPT Architecture

GPT (Generative Pre-trained Transformer) is based on the Transformer architecture. It uses self-attention mechanisms to understand relationships between words in a sequence.

Text is generated using tokenization, where text is split into tokens. Each token is assigned a probability, and the model predicts the next token step-by-step using these probabilities until a full sequence is generated.

```python
# Load GPT-2 model and tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

print("GPT-2 loaded successfully")
```

```python
# Text Generation Function

def generate_text(prompt, max_length=80, temperature=0.8, top_k=40, top_p=0.9):
    inputs = tokenizer(prompt, return_tensors="pt")

    outputs = model.generate(
        inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        max_length=max_length,
        do_sample=True,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        pad_token_id=tokenizer.eos_token_id
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
# Exercise 1: Basic Text Generation

print(generate_text("Alice was beginning"))
```

```python
# Exercise 2: Output Length Control

print("Short output:\n")
print(generate_text("The rabbit ran", max_length=30))

print("\nLong output:\n")
print(generate_text("The rabbit ran", max_length=120))
```

```python
# Exercise 3: Temperature Control

print("Low temperature (more predictable):")
print(generate_text("Once upon a time", temperature=0.3))

print("\nHigh temperature (more creative):")
print(generate_text("Once upon a time", temperature=1.2))
```

```python
# Exercise 4: Top-k and Top-p

print(generate_text("The queen said", top_k=20, top_p=0.7))
```

```python
# Exercise 5: Prompt Engineering

print(generate_text("Write a motivational message about success"))
print()
print(generate_text("Write a scary story opening"))
```

### Application Demonstration

This section demonstrates how GPT can be used for real-world content creation tasks such as writing blog introductions and creative text generation.

```python
# Application Demonstration

prompt = "Write a short blog introduction about artificial intelligence"
print(generate_text(prompt, max_length=100))
```

### Findings

The model produces coherent text but can sometimes generate repetitive or irrelevant outputs. Lower temperature values produce more structured responses, while higher values increase creativity but reduce consistency. Top-k and top-p sampling help control randomness and improve output quality.
